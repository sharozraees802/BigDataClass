### Target Encoding

```python
df2 = df
```

```python
sorted(df2["job"].unique())
```

```python
df2["job"].value_counts()
```

```python
cols_nums = {"job": {'ANALYST': 3, 'CLERK': 5, 'MANAGER': 2, 'PRESIDENT': 1, 'SALESMAN': 4} }
```

```python
df2 = df2.replace(cols_nums)
df2.head()
```

```python
df.head()
```

```python
def salCatg(salary):
    if salary >= 3000: r = 1
    elif salary >= 1500: r = 2
    else: r = 3
    return r
```

```python
df2['sal'] = df2['sal'].apply(salCatg)
df2.head()
```

### Label Encoding

```python
df3 = df[df.columns]
df3.dtypes
```

```python
df3["job"] = df3["job"].astype('category')
df3.dtypes
```

```python
sorted(df3["job"].unique())
```

```python
df3.head(50)
```

```python
df3["job"] = df3["job"].cat.codes
df3.head(50)
```

```python
df3["comm"] = df3["comm"].astype('category')
df3.dtypes
```

```python
df3["comm"] = df3["comm"].cat.codes
df3.head(50)
```

```python
df3.to_csv('D://DataSets//emp_encoded.csv', index = False)
```
